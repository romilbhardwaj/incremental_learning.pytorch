```python
def get_old_sample_weight_tensor(targets):
    train_loader = loader
    samples_trained_dict = {3: 1000, 1:2250}
    new_targets = targets.clone()
    old_targets = 0.5*targets.clone()
    
    # Get class counts in the train loader
    class_ids, counts = np.unique(train_loader.dataset.y, return_counts=True)
    new_sample_counts = dict(zip(class_ids, counts))
    
    # Get classes from targets
    _, class_tensor = targets.max(1)
    
    class_weights = {}
    for c in np.unique(class_tensor):
        class_weights[c] = get_old_sample_weight(samples_trained_dict.get(c, 0), new_sample_counts.get(c, 0))
    
    # Map
    weight_list = [class_weights[c.item()] for c in class_tensor]
    old_sample_weight_tensor = iterable_to_tensor(weight_list, 'cpu:0').unsqueeze(1)
    print(old_sample_weight_tensor)
    
    new_targets[..., :] = old_sample_weight_tensor * old_targets + (1 - old_sample_weight_tensor) * targets[..., :]
    
    return new_targets
    
def get_old_sample_weight(num_old_samples, num_new_samples):
    def func(x):
        return 1-(1/(1+(1/x)))

    if num_old_samples == 0:
        return 0
    elif num_new_samples == 0:
        raise ValueError("num_new_samples is 0, this does not make sense. Why are you calling this method?")
    else:
        ratio = num_new_samples/num_old_samples
        return func(ratio)

def iterable_to_tensor(l, device):
    return torch.from_numpy(np.array(list(l), dtype='double')).float().to(device)
```
